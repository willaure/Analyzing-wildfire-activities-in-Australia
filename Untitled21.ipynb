```python
# Practice Tasks
#TASK 1.1: Let's try to understand the change in average estimated fire area over time
#(use pandas to plot)
```

```python
# TASK 1.1: Plot change in average estimated fire area over time

# First convert 'Date' column to datetime if not already done
df['Date'] = pd.to_datetime(df['Date'])

# Group by date and calculate mean fire area
daily_mean_area = df.groupby('Year')['Estimated_fire_area'].mean()

# Create the plot
daily_mean_area.plot(
    figsize=(12, 6),
    title='Average Estimated Fire Area Over Time',
    xlabel='Year',
    ylabel='Average Fire Area (kmÂ²)',
    grid=True
)

# Improve date formatting on x-axis
plt.gcf().autofmt_xdate()
plt.tight_layout()
plt.show()
```

```python
# TASK 1.2: You can notice the peak in the plot between 2010 to 2013. Let's narrow down our finding, by plotting the estimated fire area for year grouped together with month.Â¶
```

```python
# TASK 1.2: Analyze peak fire activity (2010-2013) with monthly grouping

# 1. Filter data for 2010-2013 period
peak_years = df[(df['Date'].dt.year >= 2010) & (df['Date'].dt.year <= 2013)]

# 2. Create Year-Month groups
monthly_fire = (peak_years.groupby([peak_years['Date'].dt.year.rename('Year'),
                                  peak_years['Date'].dt.month.rename('Month')])
                ['Estimated_fire_area'].mean()
                .unstack(level=0))  # Creates columns for each year

# 3. Plot the monthly trends
plt.figure(figsize=(14, 7))
monthly_fire.plot(marker='o', linewidth=2, markersize=8)

# 4. Customize the plot
plt.title('Monthly Average Fire Area During Peak Years (2010-2013)', pad=20)
plt.xlabel('Month')
plt.ylabel('Average Fire Area (kmÂ²)')
plt.xticks(range(1,13), ['Jan','Feb','Mar','Apr','May','Jun',
                        'Jul','Aug','Sep','Oct','Nov','Dec'])
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(title='Year')
plt.tight_layout()

# 5. Highlight peak months
max_fire = monthly_fire.max().max()
plt.annotate(f'Peak Activity: {max_fire:.1f} kmÂ²',
             xy=(monthly_fire.stack().idxmax()[1], max_fire),
             xytext=(10, max_fire*0.8),
             arrowprops=dict(facecolor='red', shrink=0.05))

plt.show()
```

```python
 # Grouping the data by both 'Year' and 'Month', and calculating the mean of 'Estimated_fire_area'
df_new = df.groupby(['Year','Month'])['Estimated_fire_area'].mean()
    # Plotting the data
df_new.plot(x=df_new.index, y=df_new.values)
plt.xlabel('Year, Month')
plt.ylabel('Average Estimated Fire Area (kmÂ²)')
plt.title('Estimated Fire Area over Time')
plt.show()
```

```python
# Creating a bar plot using seaborn to visualize the distribution of mean estimated fire brightness across regions

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

# Using seaborn's barplot function to create the plot
sns.barplot(data=df, x='Region', y='Mean_estimated_fire_brightness')

# Adding labels and title
plt.xlabel('Region', fontsize=12)
plt.ylabel('Mean Estimated Fire Brightness (Kelvin)', fontsize=12)
plt.title('Distribution of Mean Estimated Fire Brightness Across Regions', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)  # In case region names are long

# Display the plot
plt.tight_layout()
plt.show()

```

```python
#TASK 1.4: Let's find the portion of count of pixels for presumed vegetation fires vary across regions
#we will develop a pie chart for this
```

```python
# Creating a pie chart to visualize the portion of count of pixels for presumed vegetation fires across regions
plt.figure(figsize=(10, 6))

# Grouping the data by region and summing the counts
region_counts = df.groupby('Region')['Count'].sum()

# Creating the pie chart
plt.pie(region_counts,
        labels=region_counts.index,
        autopct='%1.1f%%',
        startangle=140,
        shadow=True,
        textprops={'fontsize': 12})

plt.title('Percentage of Pixels for Presumed Vegetation Fires by Region', fontsize=14, fontweight='bold')
plt.axis('equal')  # Ensures pie is a circle
plt.tight_layout()
plt.show()
```

```python
# TASK 1.5: Pie chart without overlapping labels, with percentages shown in legend
import matplotlib.pyplot as plt

# Set up the figure
plt.figure(figsize=(10, 6))

# Grouping the data by region and summing the counts
region_counts = df.groupby('Region')['Count'].sum()

# Creating the pie chart without autopct to avoid overlap
patches, texts = plt.pie(region_counts,
                         labels=region_counts.index,
                         startangle=140,
                         shadow=True,
                         textprops={'fontsize': 12})

# Add title
plt.title('Percentage of Pixels for Presumed Vegetation Fires by Region', fontsize=14, fontweight='bold')

# Create legend with calculated percentages
percentages = [(i, round(k / region_counts.sum() * 100, 2)) for i, k in zip(region_counts.index, region_counts)]
legend_labels = [f"{region}: {pct}%" for region, pct in percentages]
plt.legend(patches, legend_labels, title="Region (% of Total)", loc='best', fontsize=10)

# Keep pie chart circular
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
#TASK 1.9: Let's mark these seven regions on the Map of Australia using Folium

#we have created a dataframe for you containing the regions, their latitudes and longitudes.
#For australia use [-25, 135] as location to create the map
```

```python
import folium
import pandas as pd

# Region data
region_data = {
    'region': ['NSW', 'QL', 'SA', 'TA', 'VI', 'WA', 'NT'],
    'Lat': [-31.8759835, -22.1646782, -30.5343665, -42.035067, -36.5986096, -25.2303005, -19.491411],
    'Lon': [147.2869493, 144.5844903, 135.6301212, 146.6366887, 144.6780052, 121.0187246, 132.550964]
}
reg = pd.DataFrame(region_data)

# Create a Folium map centered on Australia
australia_map = folium.Map(location=[-25, 135], zoom_start=4)

# Add region markers to the map
for i, row in reg.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Lon']],
        popup=row['region'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(australia_map)

# Display the map
australia_map
```
